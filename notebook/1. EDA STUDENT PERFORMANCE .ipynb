#### 1) Problem Statement

* The objective of this project is to develop a machine learning model that can predict the academic performance of students based on various factors, such as gender, race/ethnicity, parental education level, lunch, and test preparation course. The model should be able to accurately predict a student's performance(test scores) in terms of their grades or scores in different subjects.

* The project aims to help educators and policymakers identify students who are at risk of academic failure and provide appropriate interventions to improve their academic outcomes. Additionally, the model can be used to identify factors that have the most significant impact on student performance, which can help educators tailor their teaching methods to better meet the needs of their students.
